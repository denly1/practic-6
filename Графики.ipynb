```python
games = pd.read_csv("Video Games Sales.csv")

plt.figure(figsize=(10, 5))

plt.plot(games['Year'].head(20), games['North America'].head(20))
plt.title('North America Sales by Year')
plt.xlabel('Year')
plt.ylabel('North America Sales')
plt.show()
```

Создание линейного графика продаж видеоигр в Северной Америке по годам

На графике изображены продажи в Северной Америке по годам. Для построения графика использовалась функция plt.plot(), где по оси x отмечены годы,а по оси y  продажи.График показывает, как изменялись продажи видеоигр в Северной Америке за последние 20 лет. Из графка видно, что с 2000 по 2020 годы продажи постепенно росли.

```python
plt.plot(games['Year'].head(20), games['Europe'].head(20))
plt.title('Europe Sales by Year')
plt.xlabel('Year')
plt.ylabel('Europe Sales')
plt.show()
```

Создание линейного графика продаж видеоигр в Европе по годам

```python
plt.plot(games['Year'].head(20), games['North America'].head(20), label='North America')
plt.plot(games['Year'].head(20), games['Europe'].head(20), label='Europe')
plt.title('North America and Europe Sales by Year')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

Создание линейного графика продажи в Северной Америке и Европе по годам

```python
games = pd.read_csv('Video Games Sales.csv') 

genre_counts = games['Genre'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(genre_counts, labels=genre_counts.index, autopct='%1.1f%%')
plt.title('Жанры')
plt.show()
```

На графике представлена круговая диаграмма,отображающая распределение жанров видеоигр.Для этого была создана переменная genre_counts, в которую записаны значения количества игр каждого жанра из столбца 'Genre'. Затем с помощью функции plt.pie() была построена круговая диаграмма с метками жанров и процентным соотношением. График позволяет увидеть, какие жанры наиболее популярны среди видеоигр, причем большая часть игр приходится на жанры Action, Sports и Shooter.

Создание круговой диаграммы для столбца "genre".

```python
games = pd.read_csv('Video Games Sales.csv')
platform_counts = games['Platform'].head(50).value_counts()
plt.figure(figsize=(7, 7))
plt.pie(platform_counts, labels=platform_counts.index, autopct='%1.1f%%')
plt.title('Платформы')
plt.show()
```

График показывает распределение платформ видеоигр. Для этого была создана переменная platform_counts, в которую записаны значения количества игр на каждой платформе из столбца 'Platform'. Затем с помощью функции plt.pie() была построена круговая диаграмма с метками платформ и процентным соотношением. График позволяет увидеть, какие платформы имеют большую долю рынка видеоигр, причем большую часть приходится на платформы PlayStation, Xbox и Wii.

```python
games = pd.read_csv('Video Games Sales.csv')

sales = games.groupby('Genre').sum()[['North America', 'Europe', 'Japan', 'Rest of World']]

sales.plot(kind='bar', stacked=True)

plt.xlabel('Жанр')
plt.ylabel('Продажи (в миллионах)')
plt.show()
```

Создание столбчатой диаграммы продаж игр в North America, Europe, Japan, Rest of World

```python
games = pd.read_csv("Video Games Sales.csv")  

x = games["Game Title"].head(20)
y1 = games["North America"].head(20)
y2 = games["Europe"].head(20)
y3 = games["Japan"].head(20)

plt.figure(figsize=(29, 19))
plt.bar(x, y1, label="North America", color="blue")
plt.bar(x, y2, label="Europe", color="red")
plt.bar(x, y3, label="Japan", color="green")

plt.xlabel("Game Title")
plt.ylabel("Sales")
plt.title("Video Game Sales by Region")

plt.legend()

plt.show()
```

Создание увеличенной диаграммы рассеяния продаж игр по годам

```python
games = pd.read_csv("Video Games Sales.csv")

plt.figure(figsize=(15, 10))

plt.subplot(1, 3, 1)
plt.hist(games['Year'].head(20).dropna(), bins=20, color='skyblue')
plt.xlabel('YEAR')
plt.ylabel('Частота')
plt.title('Распределение по годам')

plt.subplot(1, 3, 2)
plt.hist(games['Genre'].head(20).dropna(), bins=games['Genre'].nunique(), color='lightgreen')
plt.xlabel('GENRE')
plt.ylabel('Частота')
plt.title('Распределение жанров')

plt.subplot(1, 3, 3)
plt.hist(games['Platform'].head(20).dropna(), bins=games['Platform'].nunique(), color='pink')
plt.xlabel('PLATFORM')
plt.ylabel('Частота')
plt.title('Распространение платформ')

plt.tight_layout()
plt.show()
```
